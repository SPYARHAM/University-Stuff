```python
import requests
from bs4 import BeautifulSoup
import csv

# Function to scrape data from Amazon
def scrape_amazon_data(category, subcategory):
    url = f"https://www.amazon.in/Pigeon-Amaze-Plus-1-5-Ltr/dp/B07WMS7TWB/ref=sr_1_1?qid=1679135585&s=appliances&sr=1-1"

    # Send a GET request to the URL
    response = requests.get(url)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find the relevant elements on the page
    product_titles = soup.find_all('span', class_='a-size-base-plus a-color-base a-text-normal')
    product_prices = soup.find_all('span', class_='a-offscreen')
    product_links = soup.find_all('a', class_='a-link-normal a-text-normal', href=True)

    # Create a list to store the scraped data
    data = []

    # Iterate over the elements and extract the desired information
    for title, price, link in zip(product_titles, product_prices, product_links):
        data.append({
            'Category': category,
            'Subcategory': subcategory,
            'Product Title': title.text.strip(),
            'Price': price.text.strip(),
            'Link': "https://www.amazon.com" + link['href']
        })

    return data

# Example usage
category = "Electronics"
subcategory = "Headphones"
data = scrape_amazon_data(category, subcategory)

# Write the data to a CSV file
filename = f"{category}_{subcategory}_data.csv"
if data:
    with open(filename, 'w', newline='', encoding='utf-8') as file:
        writer = csv.DictWriter(file, fieldnames=data[0].keys())
        writer.writeheader()
        writer.writerows(data)
    print(f"Data scraped and saved to {filename}.")
else:
    print("No data found. Unable to save to CSV.")
```

```python
import requests
from bs4 import BeautifulSoup

# Function to scrape product details from Amazon India
def scrape_amazon_india_product_details(product_link):
    # Send a GET request to the product link
    response = requests.get(product_link)

    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Extract the price
    price_element = soup.find('span', class_='a-price-whole')
    price = price_element.text.strip() if price_element else "N/A"

    # Extract the item weight
    weight_element = soup.find('th', string='Item Weight')
    weight = weight_element.find_next('td').text.strip() if weight_element else "N/A"

    # Extract the item dimensions
    dimensions_element = soup.find('th', string='Item Dimensions LxWxH')
    dimensions = dimensions_element.find_next('td').text.strip() if dimensions_element else "N/A"

    # Extract the generic name
    generic_name_element = soup.find('th', string='Generic Name')
    generic_name = generic_name_element.find_next('td').text.strip() if generic_name_element else "N/A"

    # Extract the model name
    model_name_element = soup.find('th', string='Model Name')
    model_name = model_name_element.find_next('td').text.strip() if model_name_element else "N/A"

    # Create a dictionary to store the product details
    product_details = {
        'Price': price,
        'Item Weight': weight,
        'Item Dimensions': dimensions,
        'Generic Name': generic_name,
        'Model Name': model_name
    }

    return product_details


# Example usage
product_link = "https://www.amazon.in/Pigeon-Amaze-Plus-1-5-Ltr/dp/B07WMS7TWB/ref=sr_1_1?qid=1679135585&s=appliances&sr=1-1"

product_details = scrape_amazon_india_product_details(product_link)

# Print the product details
print("Product Details:")
for key, value in product_details.items():
    print(f"{key}: {value}")
```
